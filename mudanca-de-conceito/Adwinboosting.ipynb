```python
def ABC(X, y, window = 100):
    model = ensemble.ADWINBoostingClassifier(
        model=tree.HoeffdingAdaptiveTreeClassifier(
            grace_period=50,
            delta=1e-5,
            leaf_prediction='nb',
            nb_threshold=10,
            seed=0
        ),
        n_models=30,
        seed=42
    )
    Y_pred = list()
    metric = metrics.F1()
    for xi, yi in stream.iter_pandas(X, y):
        # Test the current model on the new "unobserved" sample
        yi_pred = model.predict_one(xi)
        # Update the running metric with the prediction and ground truth value
        metric.update(yi, yi_pred)  
        # Train the model with the new sample
        model.learn_one(xi, yi)
        Y_pred.append(yi_pred)
    return(Y_pred)

 
```

## df_sudden_dos

```python
y = 1 - pd.get_dummies(df_sudden_dos.label, drop_first= True, dtype=int).benign
X = df_sudden_dos.drop(columns="label")
```

```python
Y_pred = ABC(X, y)
df = pd.DataFrame(list(zip(y, Y_pred)), columns =['y', 'y_pred'])
df.to_csv("df_sudden_dos_ABC_predict.csv",index= False, sep = ";")
```
